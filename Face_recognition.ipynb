```python
import face_recognition
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = face_recognition.load_image_file("/content/st.jpeg")
img_en = face_recognition.face_encodings(img)[0]
know = [img_en]
know_name = ['Steyn ']
```

```python
img1 = face_recognition.load_image_file("/content/ss1.jpeg")
img1_ec = face_recognition.face_encodings(img1)[0]
```

```python
 name = know_name[0]
 (top,right,bottom,left) = face_recognition.face_locations(img1)[0]
 cv2.rectangle(img1,(left,top),(right,bottom),(0,0,255),2)
 cv2.putText(img1,name,(left+2,bottom+20),cv2.FONT_HERSHEY_PLAIN,1,(255,255,255),1)
 cv2_imshow(img1)
```
